```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 10)  # 10 evenly spaced values between 0 and 10
y = np.sin(x)               # Sin function for the y-values

# Create a stem plot
plt.stem(x, y, basefmt="C0-")  # basefmt sets the style of the baseline
plt.title("Basic Stem Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
faculties = ['CSE', 'ICT', 'Physics', 'English', 'Accounting']
num_students = [120, 150, 90, 100, 100]

 # Create the bar chart with customization
plt.bar(faculties, num_students, color=['blue', 'green', 'red', 'purple', 'orange'], edgecolor='black')

# Add titles and labels
plt.title('Number of Students in Different Faculties', fontsize=16)
plt.xlabel('Faculties', fontsize=12)
plt.ylabel('Number of Students', fontsize=12)

# Annotate the bars with values
for i, v in enumerate(num_students):
    plt.text(i, v + 2, str(v), ha='center', fontsize=10)

# Display the chart
plt.show()
```

```python
# Data for two years
years = ['2023', '2024']
students_2023 = [110, 85, 60, 95, 45]
students_2024 = [120, 90, 70, 100, 50]

import numpy as np

# Set the positions and width for the bars
x = np.arange(len(faculties))
width = 0.35

# Create grouped bar chart
plt.bar(x - width/2, students_2023, width, label='2023', color='skyblue')
plt.bar(x + width/2, students_2024, width, label='2024', color='hotpink')

# Add titles, labels, and legend
plt.title('Student Enrollment by Faculty Over Years')
plt.xlabel('Faculties')
plt.ylabel('Number of Students')
plt.xticks(x, faculties)
plt.legend()

# Display the chart
plt.show()
```
